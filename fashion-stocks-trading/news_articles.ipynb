```python
import pandas as pd
from yahoo_fin import news
from datetime import datetime
from textblob import TextBlob
```

```python
stocks = ["MC.PA", "RMS.PA", "CDI.PA", "CFR.SW","LULU", "MONC.MI", "UHR.SW", "VFC", "PNDORA.CO", "RL"]
```

```python
all_news = []
for stock in stocks:
    stock_news = news.get_yf_rss(stock)
    for new in stock_news:
        all_news.append({
            'title': new.get('title', ''),
            'summary': new.get('summary', ''),
            'href': new.get('link', ''),
            'published': new.get('published', ''),
            'stock': stock
        })

# Create a DataFrame from the all_news list
news_df = pd.DataFrame(all_news)

# Save the DataFrame to a CSV file
news_df.to_csv('stock_news.csv', index=False)
```

```python
data = pd.read_csv("stock_news.csv")
```

```python
display(data)
```

```python
def reformat_date(date_string, input_format="%a, %d %b %Y %H:%M:%S %z", output_format="%d/%m/%Y"):
    date_obj = datetime.strptime(date_string, input_format)
    return date_obj.strftime(output_format)
```

```python
data['published'] = data['published'].apply(reformat_date)
```

```python
display(data)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
vader_analyzer = SentimentIntensityAnalyzer()
def get_vader_compound_sentiment(text):
    sentiment_scores = vader_analyzer.polarity_scores(text)
    return sentiment_scores['compound']
```

```python
data['title_sentiment_VADER'] = data['title'].apply(get_vader_compound_sentiment)
data['summary_sentiment_VADER'] = data['summary'].apply(get_vader_compound_sentiment)
```

```python
display(data)
```

```python
from transformers import pipeline
```

```python
def transformer_sentiment_label(text):
    nlp = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
    return nlp(text)[0]["label"]

def transformer_sentiment_score(text):
    nlp = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
    return nlp(text)[0]["score"]
```

```python
data['summary_sentiment_transformer_label'] = data['summary'].apply(transformer_sentiment_label)
data['summary_sentiment_transformer_scoe'] = data['summary'].apply(transformer_sentiment_score)
```
