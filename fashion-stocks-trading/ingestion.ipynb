# Ingesting the Datasets

```python
import pandas_datareader.data as pdr
import yfinance as yfin
import pandas as pd
import datetime as dt
```

### MC.PA

```python
yfin.pdr_override()
start = str(dt.date(2018, 1, 1))
end = str(dt.date.today())
df_mcpa = pdr.get_data_yahoo('MC.PA', start=start, end=end)
df_mcpa.head()
```

```python
#Saving the stock in data folder
df_mcpa.to_csv('data\\ingested\\mcpa.csv')
```
